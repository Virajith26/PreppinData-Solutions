```python
import pandas as pd
import numpy as np
```

```python
targets = pd.read_csv('/content/Targets.csv')
targets
```

```python
f = pd.read_csv('/content/PD 2023 Wk 1 Input.csv')
f
```

Filter the transactions to just look at DSB

```python
t1 = pd.DataFrame(list(f['Transaction Code'].str.split(pat="-")),columns=['Bank','c1','c2','c3'])
f = pd.concat([f,t1['Bank']],axis=1)
f
```

```python
f1 = f[f['Bank'] == 'DSB'].reset_index()
f1 = f1.drop(['index'],axis=1)
f1
```

Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values

```python
f1['Online or In-Person'] = f1['Online or In-Person'].apply(
    lambda x: 'Online' if x == 1 else 'In-Person'
)
f1
```

Change the date to be the quarter

```python
f1['Transaction Date'] = pd.to_datetime(f1['Transaction Date'])
```

```python
f1['Quater'] = f1['Transaction Date'].dt.to_period('Q')
f1
```

```python
t2 = f1['Quater'].astype('str').str.split(pat='Q')
t2
```

```python
Quater_number = [t2[i][1] for i in range(len(t2))]
f1['Quater No'] = Quater_number
f1
```

Sum the transaction values for each quarter and for each Type of Transaction (Online or In-Person)

```python
f2 = f1.groupby(['Quater No','Online or In-Person']).sum().reset_index()
f2
```

```python
f2 = f2.loc[:,['Quater No','Online or In-Person','Value']]
f2
```

```python
f2 = f2.astype({'Quater No': int})
f2.info()
```

For the targets: Pivot the quarterly targets so we have a row for each Type of Transaction and each Quarter

Did the pivot using Excel

```python
targets2 = pd.read_csv('/content/Targets2.csv')
targets2
```

```python
targets2.info()
```

Join the two datasets together

```python
t3 = f2.join(targets2.set_index('Quater No'),lsuffix='_T', rsuffix='_SC',on='Quater No')
t3
```

```python
t4 = t3[t3['Online or In-Person_T'] == t3['Online or In-Person_SC']]
t4
```

Calculate the Variance to Target for each row

```python
t4['Variance to Target'] = t4['Value'] - t4['Target']
t4
```

Output the data

```python
output = t4.loc[:,['Online or In-Person_T',
                   'Quater No',
                   'Value',
                   'Target',
                   'Variance to Target']]
output
```

```python
output.to_csv('output.csv',index=False)
```
