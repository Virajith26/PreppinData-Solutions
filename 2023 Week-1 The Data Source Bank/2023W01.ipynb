```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
f = pd.read_csv('PD 2023 Wk 1 Input.csv')
f
```

Split the Transaction Code to extract the letters at the start of the transaction code. Rename the new field with the Bank code 'Bank'

```python
t1 = pd.DataFrame(list(f['Transaction Code'].str.split(pat="-")),columns=['Bank','c1','c2','c3'])
t1
```

```python
f2 = pd.concat([f,t1['Bank']],axis = 1)
f2
```

Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values.

```python
f2['OOnline or In-Person'] = f2['Online or In-Person'].apply(
    lambda x: 'Online' if x == 1 else 'In-person'
)
f2
```

Change the date to be the day of the week

```python
f2['Transaction Only Date'] = pd.to_datetime(f2['Transaction Date']).dt.day_name()
f2
```

```python
final_f = f2.drop(['Transaction Date','Online or In-Person','Transaction Code'],axis=1)
final_f.rename(columns = {'OOnline or In-Person':'Online or In-Person',
                          'Transaction Only Date':'Transaction Date' },inplace=True)
final_f
```

OUTPUT 1: Total Values of Transactions by each bank

```python
output1 = pd.DataFrame(final_f.groupby(['Bank']).sum()).reset_index().loc[:,['Bank','Value']]
output1
```

2. Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)

```python
output2 = final_f.groupby(['Bank','Transaction Date','Online or In-Person']).sum().reset_index()
output2
```

```python
output3 = final_f.groupby(['Bank','Customer Code']).sum().reset_index()
output3
```

Printing the DataFrames:

```python
output1.to_csv('output1.csv',index=False)
output2.to_csv('output2.csv',index=False)
output3.to_csv('output3.csv',index=False)
```
